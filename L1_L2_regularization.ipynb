```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('house.csv')
```

```python
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cells_to_fill_zero = ['bedrooms']
dataset[cells_to_fill_zero] = dataset[cells_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['dfg'] = dataset['dfg'].fillna(dataset.dfg.mean())
```

```python
dataset.dropna(inplace=True)
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
x = dataset.drop('price', axis=1)
y = dataset['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg  = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

Lasso_regression >> L1 Regularization

```python
from sklearn.linear_model import Lasso
```

```python
lso = Lasso(alpha=50, max_iter=100, tol=0.1)
```

```python
lso.fit(X_train, y_train)
```

```python
lso.score(X_test, y_test)
```

```python
lso.score(X_train,y_train)
```

Ridge >> L2 regularizationn

```python
from sklearn.linear_model import Ridge
```

```python
ridg = Ridge(alpha=50, max_iter=100, tol=0.1)
ridg.fit(X_train,y_train)
```

```python
ridg.score(X_test, y_test)
```

```python
ridg.score(X_train,y_train)
```
