```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y,lr):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = lr

    %matplotlib inline
    plt.xlabel('x',fontsize = 20)
    plt.ylabel('y',fontsize = 20)
    plt.scatter(x, y, color="red", marker="*")
    plt.title(f"learning rate = {learning_rate}")

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        plt.plot(x,y_predicted, color="yellow" )
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr -= learning_rate * md
        b_curr -= learning_rate * bd
        # print(f"m->{m_curr} , b->{b_curr}, cost->{cost}, iteration->{i}")
    plt.plot(x, y, color="red")
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y,lr=0.001)
```

```python
gradient_descent(x,y,0.01)
```

```python
gradient_descent(x,y,0.05)
```
